* Removing the data where age is misleading or in accurate

```python
data["Age"].value_counts()
```

```python
data1=data[data["Age"]>6]
data1
```

* populating the state column

```python
data1[["address","state$zip"]]=data1["Address"].str.split(",",expand=True)
data1[["c","State","Zip","c1","c2"]]=data1["state$zip"].str.split(" ",expand=True)
data2=data1.drop(columns=["state$zip","c","c1","c2"])
data2["Type_reraction"]=data2["Type"]
data3=data2.drop(columns="Type")
data3
```

```python
data3.to_csv("user_data.csv")
```

#### Removing Duplicates to get User-wise Data Not traffic-wise data

```python
data4=data3.drop(columns=["Datetime","Type_reraction","Category"])
data5=data4.drop_duplicates()
data5
```
